```python
import pandas as pd
from functools import reduce

# 存储读取的数据DataFrame
df_population_density = pd.read_excel('人口密度.xlsx')
df_population_size = pd.read_excel('人口规模.xlsx')
df_urbanization_rate = pd.read_excel('城镇化率.xlsx')

employment_info_path = '就业信息.xlsx'
df_unemployment_rate = pd.read_excel(employment_info_path, '城镇失业率')
df_employment_number = pd.read_excel(employment_info_path, '从业人员数')
df_employment_industries = pd.read_excel(employment_info_path, '第一、二、三产业就业人数')

df_salary_level = pd.read_excel('工资水平.xlsx')
df_age_structure = pd.read_excel('年龄结构.xlsx').rename(columns={'城市': '城市名称'})

life_path = '生活水平.xlsx'
life_avg_in = pd.read_excel(life_path, '人均可支配收入', header=1)
life_avg_out = pd.read_excel(life_path, '人均消费支出', header=1)
life_urban_avg_in = pd.read_excel(life_path, '城镇居民人均收入', header=1)
life_urban_avg_out = pd.read_excel(life_path, '城镇居民消费支出', header=1)
life_rural_avg_in = pd.read_excel(life_path, '农村居民人均收入', header=1)
life_rural_avg_out = pd.read_excel(life_path, '农村居民消费支出', header=1)

# 数据结构转换
df_salary_level = pd.melt(df_salary_level, id_vars=['averageWage'], var_name='城市名称', value_name='平均收入')
df_salary_level.columns = ['年份', '城市名称', '平均收入']
df_salary_level['年份'] = df_salary_level['年份'].str.replace('年', '').astype('int64')
life_avg_in = pd.melt(life_avg_in, id_vars=['城市名称'], var_name='年份', value_name='人均可支配收入')
life_avg_in.columns = ['城市名称', '年份', '人均可支配收入']
life_avg_out = pd.melt(life_avg_out, id_vars=['城市名称'], var_name='年份', value_name='人均消费支出')
life_avg_out.columns = ['城市名称', '年份', '人均消费支出']
life_urban_avg_in = pd.melt(life_urban_avg_in, id_vars=['城市名称'], var_name='年份', value_name='城镇居民人均收入')
life_urban_avg_in.columns = ['城市名称', '年份', '城镇居民人均收入']
life_urban_avg_out = pd.melt(life_urban_avg_out, id_vars=['城市名称'], var_name='年份', value_name='城镇居民消费支出')
life_urban_avg_out.columns = ['城市名称', '年份', '城镇居民消费支出']
life_rural_avg_in = pd.melt(life_rural_avg_in, id_vars=['城市名称'], var_name='年份', value_name='农村居民人均收入')
life_rural_avg_in.columns = ['城市名称', '年份', '农村居民人均收入']
life_rural_avg_out = pd.melt(life_rural_avg_out, id_vars=['城市名称'], var_name='年份', value_name='农村居民消费支出')
life_rural_avg_out.columns = ['城市名称', '年份', '农村居民消费支出']


print("人口密度:")
print(df_population_density)

print("人口规模:")
print(df_population_size)

print("城镇化率:")
print(df_urbanization_rate)

print("城镇失业率:")
print(df_unemployment_rate)

print("从业人员数:")
print(df_employment_number)

print("第一、二、三产业就业人数:")
print(df_employment_industries)

print("工资水平:")
print(df_salary_level)

print("年龄结构:")
print(df_age_structure)

print("人均可支配收入:")
print(life_avg_in)

print("人均消费支出:")
print(life_avg_out)

print("城镇居民人均收入:")
print(life_urban_avg_in)

print("城镇居民消费支出:")
print(life_urban_avg_out)

print("农村居民人均收入:")
print(life_rural_avg_in)

print("农村居民消费支出:")
print(life_rural_avg_out)

dfs = [df_population_density, df_population_size, df_urbanization_rate, df_unemployment_rate, df_employment_number,
       df_employment_industries, df_salary_level, df_age_structure, life_avg_in, life_avg_out, life_urban_avg_in,
       life_urban_avg_out, life_rural_avg_in, life_rural_avg_out]
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['城市名称', '年份'], how='outer'), dfs)
print(merged_df)
merged_df.to_excel('merged_file.xlsx', index=False)
```
